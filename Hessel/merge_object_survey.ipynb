```python
import pandas as  pd
import numpy as np
object_labels = pd.read_pickle("object_labels.pickle")
object_labels.columns = object_labels.columns.str.lower()
```

```python
object_labels.head()
```

```python
object_labels['data_amz_label'].value_counts()
```

```python
dominant_objects = object_labels.loc[object_labels.groupby('image_id')['data_amz_label_confidence'].idxmax()] 
dominant_objects = dominant_objects.sort_index()
```

```python
survey = pd.read_pickle("survey.pickle")
survey.rename(columns={'insta_user_id':'user_id'}, inplace=True)
survey = survey[['user_id', 'PERMA', 'P', 'E', 'R', 'M', 'A', 'P_EMO', 'N_EMO', 'gender']]
image_data = pd.read_pickle("image_data.pickle")
survey.columns = survey.columns.str.lower()
```

```python
image_object_df = pd.merge(image_data, dominant_objects, how='inner', on='image_id')
image_object_df.head()
```

```python
objects_per_user = pd.crosstab(image_object_df['user_id'], image_object_df['data_amz_label'])
objects_per_user= objects_per_user.reset_index().rename_axis(None, axis=1)
objects_per_user = objects_per_user[['user_id', 'Plant', 'Person', 'Animal', 'Food', 'Art']]
objects_per_user
```

```python
objects_per_user['user_id'] = objects_per_user['user_id'].astype(str).astype('int64')
survey_objects_per_user = pd.merge(objects_per_user, survey, how='inner', on='user_id')
survey_objects_per_user.tail()
```

```python
dominant_object_per_user = image_object_df.groupby('user_id')['data_amz_label'].apply(lambda x: x.value_counts().index[0]).reset_index()
dominant_object_per_user.info()
```

```python
dominant_object_per_user['user_id'] = dominant_object_per_user['user_id'].astype(str).astype('int64')
```

```python
survey_dominant_object = pd.merge(dominant_object_per_user, survey, how='inner', on='user_id')
survey_dominant_object.tail()
```
