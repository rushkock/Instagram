```python
import pandas as  pd
import numpy as np
face = pd.read_pickle("face.pickle")
survey = pd.read_pickle("survey.pickle")
survey.rename(columns={'insta_user_id':'user_id'}, inplace=True)
survey = survey[['user_id', 'PERMA', 'P', 'E', 'R', 'M', 'A', 'P_EMO', 'N_EMO', 'gender']]
survey.columns = survey.columns.str.lower()
image_data = pd.read_pickle("image_data.pickle")
survey.head()
```

```python
#take subset of the dataframe with relevant information
faces_ss = face[['image_id','face_id','face_smile','face_smile_confidence','face_emo','emo_confidence']]
faces_ss.head(5)
```

```python
faces_ss.info()
```

```python
#Filter the most dominant emotion per picture
faces_ss_picture = faces_ss.loc[faces_ss.groupby('image_id')['emo_confidence'].idxmax()]        
faces_df = faces_ss_picture.sort_index()
faces_df.head()
```

```python
#Merge dataframs image_data and face subset data
image_face_df = pd.merge(image_data, faces_df, how='inner', on='image_id')
image_face_df.head()
```

```python
# Calculate the mean emotion per user
image_face_df['face_emo'].value_counts()
```

```python
df_emotions_per_user = pd.crosstab(image_face_df['user_id'], image_face_df['face_emo'])
df_emotions_per_user= df_emotions_per_user.reset_index().rename_axis(None, axis=1)
df_emotions_per_user
```

```python
df_emotions_per_user['user_id'] = df_emotions_per_user['user_id'].astype(str).astype('int64')
df_emotions_per_user.info()
```

```python
survey.info()
```

```python
survey_emotion_per_user = pd.merge(df_emotions_per_user, survey, how='inner', on='user_id')
survey_emotion_per_user.tail()
```

```python
survey_emotion_per_user.to_pickle('survey_emotion_per_user.pickle')
```

```python
image_face_df.tail()
```

```python
dominant_emotion_per_user = image_face_df.groupby('user_id')['face_emo'].apply(lambda x: x.value_counts().index[0]).reset_index()
#dominant_emotion_per_user = pd.crosstab(image_face_df['user_id'], image_face_df['face_emo'])
dominant_emotion_per_user.info()
```

```python
dominant_emotion_per_user['user_id'] = dominant_emotion_per_user['user_id'].astype(str).astype('int64')
dominant_emotion_per_user.info()
```

```python
survey_dominant_emotion = pd.merge(dominant_emotion_per_user, survey, how='inner', on='user_id')
survey_dominant_emotion.tail()
```

```python
survey_dominant_emotion.to_pickle('survey_dominant_emotion')
```
